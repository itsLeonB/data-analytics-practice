# Games Sales Data Analytics

```python
import pandas as pd
```

```python
games = pd.read_csv('Games Sales - Games.csv')
games.info()
```

```python
games.head()
```

```python
games['Release'] = pd.to_datetime(games['Release'])
games.head()
```

```python
release = games[['Name', 'Release']]
release.head()
```

Which game is the oldest?

```python
oldest_games = release.sort_values(by=['Release'])
oldest_games.head()
```

Which game is the newest?

```python
newest_games = release.sort_values(by=['Release'], ascending=False)
newest_games.head()
```

Which publisher published most of the games?

```python
games.value_counts('Publisher')
```

Which developer developed most of the games?

```python
games.value_counts('Developer')
```

```python
series = games[['Series', 'Name', 'Sales']].dropna()
series.head()
```

Which series have the most sales?

```python
most_sales_by_series = series.groupby('Series').sum()
most_sales_by_series.sort_values(by=['Sales'], ascending=False).head()
```

Which series have the most games?

```python
series.value_counts('Series')
```
